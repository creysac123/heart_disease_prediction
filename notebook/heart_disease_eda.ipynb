```python
def display_five_number_summary(df):
    # Separate continuous columns
    continuous_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Display the 5-number summary for each continuous column
    summary = {}
    for col in continuous_columns:
        summary[col] = {
            'Min': df[col].min(),
            '25th Percentile': df[col].quantile(0.25),
            'Median': df[col].median(),
            '75th Percentile': df[col].quantile(0.75),
            'Max': df[col].max()
        }
    
    # Convert the summary dictionary to a DataFrame for better readability
    summary_df = pd.DataFrame(summary).T
    
    return summary_df

# Display the 5-number summary before imputation
display_five_number_summary(df)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

def impute_missing_values(df):
    # Separate categorical and continuous features
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    continuous_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Drop continuous columns that don't have any non-missing values
    continuous_columns = [col for col in continuous_columns if df[col].notna().sum() > 0]
    
    # Impute continuous features with the median
    if len(continuous_columns) > 0:  # Only apply if there are valid continuous columns
        median_imputer = SimpleImputer(strategy='median')
        df[continuous_columns] = median_imputer.fit_transform(df[continuous_columns])
    
    # Impute categorical features with the most frequent value
    if len(categorical_columns) > 0:  # Only apply if there are valid categorical columns
        frequent_imputer = SimpleImputer(strategy='most_frequent')
        df[categorical_columns] = frequent_imputer.fit_transform(df[categorical_columns])
    
    return df

# Replace empty spaces with NaN
df.replace(" ", pd.NA, inplace=True)

# Apply the imputation function
df = impute_missing_values(df)

# Check the result
print(df.isnull().sum())  # This should show no missing values if all were imputed
```

```python
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                # looping on the number of columns
        if df[column].nunique() < 32:       # nunique:  returns unique values in the column,
                                                    #here 43 is the hyperparamter and can be changed
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         # Calling the function

# # Tabulate is a package used to print the list, dict or any data sets in a proper format; in table format
from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "Continuous": continuous}, headers = ["categorical", "continuous"]))
```

## Level 1 Analysis:

### User defined functions for level 1 analysis of categorical columns
This function will check for the following:


- unique values 
- mode 
- sum of all the null values 
- Total count of missing values

```python
def info_of_cat(col): 
    print(f"Unique values in {col} are: {df[col].unique()}") # unique: returns the unique values in the col
    print(f"Mode of {col} is {df[col].mode()[0]}")  # mode: returns the mode of the column
    print(f"Number of missing values in {col} is {df[col].isnull().sum()}") # isnull().sum() counts the number of null in dataframe
    if df[col].isnull().sum() > 0:  # check if null values are present
        print(f"\nThere are null values in the {col} column")
```

# $$\textbf{\textit{Univariate Analysis of Numerical Variables}}$$

```python
def info_of_numerical(col):
    # Print mean, median, mode, and standard deviation
    print(f"The mean of the {col} is {df[col].mean()}")
    print(f"The mode of the {col} is {df[col].mode()[0]}")
    print(f"The standard deviation of the {col} is {df[col].std()}")
    
    # Print number of missing values
    print(f"Number of missing values in the {col} is {df[col].isnull().sum()}")
    
    # Print five-number summary
    min_val = df[col].min()
    q1 = df[col].quantile(0.25)
    median = df[col].median()
    q3 = df[col].quantile(0.75)
    max_val = df[col].max()
    
    print(f"The minimum value of the {col} is {min_val}")
    print(f"The first quartile (25th percentile) of the {col} is {q1}")
    print(f"The median (50th percentile) of the {col} is {median}")
    print(f"The third quartile (75th percentile) of the {col} is {q3}")
    print(f"The maximum value of the {col} is {max_val}")
```
